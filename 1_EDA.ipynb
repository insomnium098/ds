```python
# Importando las librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from scipy import stats
from sklearn.ensemble import IsolationForest
```

### Cargando el dataset de cancer de mama 
https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic


```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Mostrando los primeros rows del dataset
print(df.head())
```

```python
# Resumen esadistico
print(df.describe())
```

```python
# Revisando NAs
print(df.isnull().sum())
```

### Visualizacion de los datos

```python
# Histogramas de cada feature
df.hist(bins=20, figsize=(20, 15))
plt.show()

# Box plots para mostrar outliers
plt.figure(figsize=(20, 10))
df.drop('target', axis=1).boxplot()
plt.xticks(rotation=45)
plt.show()
```

```python
# Matriz de correlacion
corr_matrix = df.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()

# Pairplot usando solo agunos de los features (columnas)
sample_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'target']
sns.pairplot(df[sample_features], hue='target')
plt.show()
```
